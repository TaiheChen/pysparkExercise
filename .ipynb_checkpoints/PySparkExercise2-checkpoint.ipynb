```python
#set up environment
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('data_processing').getOrCreate()
import pyspark.sql.functions as F
from pyspark.sql.types import *
```

```python
# Groupby Collect UDF Join Pivoting Rank

df = spark.read.csv("BankChurners.csv",header=True,inferSchema=True)
df.groupby('Education_Level').count().orderBy('count',ascending=False).show()
```

```python
df.groupBy('Education_Level').agg(F.avg("Customer_Age").alias('mean_age')).orderBy('mean_age',ascending=False).show()
```

```python
df.groupBy('Education_Level').agg(F.collect_set("Income_Category")).show(truncate=False)
#truncate:Full description
```

```python
df.groupBy('Education_Level').agg(F.collect_list("Income_Category")).show(truncate=False)
```

```python
from pyspark.sql.functions import udf
def salary_cat(income):
    if income == "$40K - $60K":
        return "Low"
    if income == "$60K - $80K":
        return "Mediam"
    if income == "$80K - $120K":
        return "High"
    if income == "$120K +":
        return "Very High"
    if income == "Less than $40K":
        return "Very Low"
    else:
        return "Unknown"
income_udf=udf(salary_cat, StringType())
df=df.withColumn('salary_cat',income_udf(df['Income_Category']))
df.select('Education_Level','salary_cat').show()
```

```python
df.groupBy('Education_Level').pivot('salary_cat').sum('Months_on_book').fillna(0).show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import col,row_number
win = Window.orderBy(df['Education_Level'].desc())
df=df.withColumn('rank',row_number().over(win).alias('rank'))
df.select("Education_Level","rank").show()
```
